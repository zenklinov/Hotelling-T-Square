```python
#Awal Package NN
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
#Deep Auto Encoder
### Define the encoder dimension
encoding_dim = 8
```

```python
features_tra_x = features_tra.drop(columns=['proto','service','state', 'is_ftp_login', 'is_sm_ips_ports'])
features_tes_x = features_tes.drop(columns=['proto','service','state', 'is_ftp_login', 'is_sm_ips_ports'])
```

```python
#Scaling data
from sklearn.preprocessing import MinMaxScaler

def scale_datasets(features_tra_x, features_tes_x):
    robust_scaler = MinMaxScaler()
    features_train_scaled = pd.DataFrame(
        robust_scaler.fit_transform(features_tra_x),
        columns = features_tra_x.columns
    )
    features_test_scaled = pd.DataFrame(
        robust_scaler.transform(features_tes_x),
        columns = features_tes_x.columns
    )
    return features_train_scaled, features_test_scaled

features_train_scaled, features_test_scaled = scale_datasets(features_tra_x, features_tes_x)
```

```python
features_test_scaled
```

```python
#Dimensional Reduction with NN AutoEncoder
from keras.layers import Input, Dense
from keras.models import Model
from tensorflow.keras.optimizers import Adam


#Setting Optimizer
opt = Adam(learning_rate=0.001,beta_1=0.9, beta_2=0.999, epsilon=1e-08)

input_dim = Input(shape = (37, ))

# Encoder Layers
encoded1 = Dense(30, activation = 'tanh')(input_dim)
encoded2 = Dense(20, activation = 'tanh')(encoded1)
encoded3 = Dense(encoding_dim, activation = 'tanh')(encoded2)

# Decoder Layers
decoded1 = Dense(20, activation = 'tanh')(encoded3)
decoded2 = Dense(30, activation = 'tanh')(decoded1)
decoded3 = Dense(37, activation = 'sigmoid')(decoded2)

# Combine Encoder and Deocder layers
autoencoder = Model(inputs = input_dim, outputs = decoded3)

# Compile the Model
autoencoder.compile(
    loss='mae',
    metrics=['MeanSquaredError', 'MeanAbsoluteError'],
    optimizer=opt
)

#Summary
autoencoder.summary()
```

```python
autoencoder.fit(
    features_train_scaled, 
    features_train_scaled, 
    epochs = 10,
    shuffle = False, 
    validation_data = (features_test_scaled, features_test_scaled)
)
```
